##OBESERVATION:

## From the above null values we know that for each category other cateregeries values(column) are null so we will remove those feature for particular dataframe

###Eg. for race total category other categories(age, education, gender, income, race) are null.

## Dropping Missing values
###(According to above observation)

```python
columns_to_drop = ['income','age(years)','gender','education','total']
race.drop(columns=columns_to_drop, inplace=True)
columns_to_drop = ['race/ethnicity','age(years)','gender','education','total']
inc.drop(columns=columns_to_drop, inplace=True)
columns_to_drop = ['race/ethnicity','income','gender','education','total']
age.drop(columns=columns_to_drop, inplace=True)
columns_to_drop = ['race/ethnicity','income','age(years)','gender','total']
edu.drop(columns=columns_to_drop, inplace=True)
columns_to_drop = ['race/ethnicity','income','age(years)','education','total']
gen.drop(columns=columns_to_drop, inplace=True)
columns_to_drop = ['race/ethnicity','income','age(years)','gender','education']
total.drop(columns=columns_to_drop, inplace=True)
```

```python
dataframes = [race,inc,age,edu,gen,total]
```

```python
for i, df in enumerate(dataframes, start=1):
    null_count = df.isnull().sum().sum()
    print(f"Null values in df{i}: {null_count}")
```
